```python
#show distribution of movie release years
df['release_year'].plot(kind='hist', title='release year distribution')
```

```python
#show distribution of movie durations in minutes
df['runtime'].plot(kind='hist', title='duration distribution')
```

```python
#show distribution of average ratings
df['vote_average'].plot(kind='hist', title='rating distribution')
```

```python
#line chart showing how the number of movie releases changed over the years
df['release_year'].value_counts().sort_index().plot(kind='line', title='movies released over the years')
```

```python
#Convert release_date to datetime format to extract months or years
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

```python
#Bar chart showing how many movies were released in each month
df['release_date'].dt.month.value_counts().sort_index().plot(kind='bar', title='releases by month')
```

```python
#line chart to show the trend of movie releases over the years
df['release_year'].value_counts().sort_index().plot(kind='line', title='movies released over time')
```

```python
#Histogram to show how movie ratings are distributed
df['vote_average'].plot(kind='hist', title='distribution of ratings')
```

```python
#Scatter plot to show the relationship between movie budget and revenue
df.plot(kind='scatter', x='budget', y='revenue', title='budget vs revenue')
```

#Summary of Key Findings:
	1.	Movie Release Trends:
	•	The number of movies released increased steadily over the years, especially after 2000.
	•	Peak release periods were often observed in July and December.
	2.	Ratings Distribution:
	•	Most movies have an average rating between 6 and 8, indicating generally favorable reviews.
	3.	Duration Distribution:
	•	The typical movie duration falls between 90 and 120 minutes, with a few outliers on both ends.
	4.	Budget and Revenue:
	•	There is a positive relationship between budget and revenue — higher-budget films tend to earn more.
	5.	Release Date Patterns:
	•	Certain months (like July and December) had more movie releases, possibly due to summer and holiday demand.
	6.	Popularity Levels:
	•	Movies were fairly evenly distributed across categories like High, Medium, Moderately High, and Low popularity.
